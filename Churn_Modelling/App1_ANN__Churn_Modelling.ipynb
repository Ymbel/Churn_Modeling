2.   les pratiques à suivre pour préparer les données sont:

*   L 'Exploratoire des Données: Comprendre la structure des données, détecter les tendances, les outliers, les distributions, et les relations entre les variables.
*   Gestion des Valeurs Manquantes: Les valeurs manquantes peuvent entraîner des erreurs lors de l'entraînement du modèle. C'est là où intervient le data scientist pour trouver la meilleur stratégie (suppression, imputation(moyenne, medianne...))
*   Gestion des Outliers: de même pour les outliers, ils peuvent entraîner des erreurs lors de l'entraînement du modèle. là on doit doit les supprimer ou les imputer(moyenne, medianne...).
*   Encodage des Variables Catégorielles:  Les algorithmes de machine learning nécessitent souvent des données numériques. il y'a pas mal de méthode pour encoder les variables categorielles (c'est ce qu'on va voir à travers ce mini_projet).
*   Normalisation ou Standardisation: la normalisation et la standarisation des données jouent un rôle crutiale lors de la preparation des données et aussi dans l'optimisation du modèle.
*   Validation Croisée : la validation croisée pour réduire le risque de surajustement aux données d'entraînement.
3.   L'architecture du réseaux de neurons:
Input Layer (Taille de l'entrée) -> Hidden Layer (ReLU) -> Hidden Layer (ReLU) -> Output Layer (Sigmoid)

4.  Les fonctions d’activation à choisir sont:


*   RELU: géneralemment utilisée dans les couche cachées d'un réseau de neurons. Il permet de mieux apprendre des motifs complexes
*   Sigmoid: cette fonction est utilisée dans la couche de sortie d'un réseau de neurones pour des tâches de classification binaire.

5.
En apprentissage automatique, la généralisation est la capacité d'un modèle à effectuer des prédictions précises sur des données qu'il n'a pas vues au cours de l'entraînement.
L'une des pratiques les plus courantes pour assurer la généralisation d'un modèle est la validation croisée.
La régularisation est une méthode qui ajoute une pénalité au modèle, ce qui l'empêche de s'adapter trop aux données d'entraînement.

6.

*   L'overfitting se produit lorsque le modèle est trop complexe et qu'il apprend les détails spécifiques des données d'entraînement. Cela peut entraîner des performances médiocres sur les données de test,car le modèle ne sera pas en mesure de généraliser à de nouvelles données qui ne présentent pas les mêmes détails spécifiques.
*   L'underfitting se produit lorsque le modèle est trop simple et qu'il ne parvient pas à capturer la relation entre les données d'entrée et de sortie. Cela peut entraîner des performances médiocres sur les données d'entraînement et les données de test, car le modèle ne sera pas en mesure de faire des prédictions précises.

7.


*   Le biais est une mesure de la précision du modèle sur les données d'entraînement. Un modèle avec un faible biais est un modèle qui est précis sur les données d'entraînement.
*   La variance est une mesure de la variation des prédictions du modèle sur de nouvelles données. Un modèle avec une faible variance est un modèle qui produit des prédictions similaires sur de nouvelles données.

Un modèle avec un faible biais et une faible variance sera un modèle précis sur les données d'entraînement et les données de test

8. La k-fold cross-validation consiste à diviser l'ensemble de données d'entraînement en k sous-ensembles, appelés folds. Le modèle est ensuite entraîné sur k-1 folds et évalué sur le fold restant. Ce processus est répété k fois, en utilisant un fold différent comme ensemble de test à chaque fois.

9. dans un probleme de classification on veut bien que le modèle performe bien non seulement sur les données d'entrainement (seen and known data ) mais aussi sur les donnèes du test (unseen and unknown data), cad qu'on doit les géneraliser même sur les données de test


*   Comment le k-fold cross validation peut résoudre le problème d’overfitting:
Cross validation est une technique pour résoudre le problème de l'overfitting, Lorsque le modèle est entraîné sur k-1 folds, il ne voit pas le fold restant. Lors de l'évaluation du modèle sur ce fold restant, il ne peut pas se baser sur les données qu'il a déjà vues. Cela permet de tester la capacité du modèle à faire des prédictions précises sur des données qu'il n'a pas vues auparavant.
*  Comment peut assurer un Bias variance tradefoff: La validation croisée fournit une estimation plus fiable de la performance du modèle en moyennant les résultats sur plusieurs ensembles de validation.
Si le modèle a un taux d'erreur élevé sur le fold de test, cela signifie qu'il est trop complexe et qu'il est susceptible de surapprendre.Inversement, si le modèle a un taux d'erreur faible sur le fold de test, cela signifie qu'il est trop simple et qu'il est susceptible de sous-apprendre.

```python
# Removing the unnecassary features from the dataset

data = data.drop(['CustomerId', 'Surname', 'RowNumber'], axis = 1)



print(data.columns)
```

```python

data.shape
```

```python
# splitting the dataset into x(independent variables) and y(dependent variables)

x = data.iloc[:,0:10]
y = data.iloc[:,10]

print(x.shape)
print(y.shape)

print(x.columns)
#print(y)
```

```python
x = pd.get_dummies(x)

x.head()
```

```python
x.columns
```

```python

# splitting the data into training and testing set

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python

# Feature Scaling
# Only on Independent Variable to convert them into values ranging from -1 to +1

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)

x_train = pd.DataFrame(x_train)
x_train.head()
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python

# creating the model
model = Sequential()

# first hidden layer
model.add(Dense(8, activation = 'relu', input_shape=(x_train.shape[1],)))

# second hidden layer
model.add(Dense(8, activation = 'relu'))

# third hidden layer
model.add(Dense(8, activation = 'relu'))

# fourth hidden layer
model.add(Dense(8, activation = 'relu'))

# fifth hidden layer
model.add(Dense(8, activation = 'relu'))

# output layer
model.add(Dense(1, activation = 'sigmoid'))
```

```python
# Compiling the NN
# binary_crossentropy loss function used when a binary output is expected
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=100, batch_size= 10)
```

```python
y_pred = model.predict(x_test)
y_pred = (y_pred > 0.5)
y_pred
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python

from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score

from keras.layers import Dense
from keras.models import Sequential

def build_classifier(optimizer):
  # creating the model
  model = Sequential()

  # first hidden layer
  model.add(Dense(8, activation = 'relu', input_shape=(x_train.shape[1],)))

  # second hidden layer
  model.add(Dense(8, activation = 'relu'))

  # output layer
  model.add(Dense(1, activation = 'sigmoid'))

  # Compiling the NN
  # binary_crossentropy loss function used when a binary output is expected
  model.compile(optimizer = optimizer, loss = 'binary_crossentropy', metrics = ['accuracy'])

  return model

model = KerasClassifier(build_fn = build_classifier, batch_size = 10, nb_epoch = 50)
```
